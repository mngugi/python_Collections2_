```python
import hashlib
import argparse
from pathlib import Path

CHUNK_SIZE = 1024 * 1024  # 1MB chunks


def compute_hash(file_path, algorithm="sha256"):
    """Compute cryptographic hash of a file using chunked reading."""
    
    if algorithm not in hashlib.algorithms_available:
        raise ValueError(f"Unsupported algorithm: {algorithm}")

    hasher = hashlib.new(algorithm)

    with open(file_path, "rb") as f:
        while chunk := f.read(CHUNK_SIZE):
            hasher.update(chunk)

    return hasher.hexdigest()


def verify_hash(file_path, expected_hash, algorithm="sha256"):
    """Verify file hash against expected value."""
    
    calculated_hash = compute_hash(file_path, algorithm)

    print(f"\nFile: {file_path}")
    print(f"Expected:  {expected_hash}")
    print(f"Calculated:{calculated_hash}")

    if calculated_hash == expected_hash:
        print("✔ Integrity verified")
        return True
    else:
        print("✘ Hash mismatch (possible tampering)")
        return False


def main():
    parser = argparse.ArgumentParser(description="Hash ML model weights securely")
    
    parser.add_argument("file", help="Model file path")
    parser.add_argument("-a", "--algorithm", default="sha256",
                        help="Hash algorithm (default: sha256)")
    parser.add_argument("-v", "--verify", help="Verify against known hash")

    args = parser.parse_args()

    file_path = Path(args.file)

    if not file_path.exists():
        print("File not found.")
        return

    if args.verify:
        verify_hash(file_path, args.verify, args.algorithm)
    else:
        hash_value = compute_hash(file_path, args.algorithm)
        print(f"{args.algorithm.upper()} hash:")
        print(hash_value)


if __name__ == "__main__":
    main()
```
