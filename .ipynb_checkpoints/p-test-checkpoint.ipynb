```python
import numpy as np
from scipy import stats

dt = np.random.normal(loc=0, scale=1,  size=30)
t_stat, p_value = stats.ttest_1samp(dt, popmean=0)

print("T-statistic: ", t_stat)
print("P-Value:", p_value)
```
