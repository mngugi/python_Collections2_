```python
import tkinter as tk
import arrayCall

root = tk.Tk()
root.title("Find Minimum Element")
root.geometry("350x350")

arr = [0.1, 0.99, 0.0999, 0.99999, 1]

def find_min_element():
    min_val = arrayCall.find_min(arr)
    result_label.config(text=f"The minimal element is: {min_val}")

# Title label
title_label = tk.Label(root, text="Minimum Value Finder", font=("Mono", 12))
title_label.pack(pady=10)

# Array display
array_label = tk.Label(root, text=f"Array: {arr}")
array_label.pack(pady=5)

#button icon

btn_icon = tk.PhotoImage(file="check.png")

# Button
min_button = tk.Button(root,image=btn_icon, text="Find Minimum",compound=tk.LEFT, command=find_min_element)
min_button.pack(pady=10)

# 
min_button.image = btn_icon   # ðŸ”‘ critical line
min_button.pack(pady=10)

# Result label
result_label = tk.Label(root, text="", font=("Mono", 10))
result_label.pack(pady=10)

# Run GUI
root.mainloop()

    
    
```
