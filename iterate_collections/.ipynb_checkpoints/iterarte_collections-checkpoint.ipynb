```python
num_list = [9,5,6,8,3,2]
mix_tuple = [78,9, 'damaged roots',25, 45]
set_characters ={'comunity', 'child', 'activity'} # we use curly brackets
strings = 'formalemploment'
dictionary = {'i': 23, 'ii': 65, 'iii': 58 }

print('------------ Lists --------------\n')

for i in num_list:
    print(i)

print('------------ Tuple --------------\n')

for item in mix_tuple:
    print(item)
    
print('----------- Items --------------------\n')

for j in set_characters:
    print(j)
print('---------- Characters ----------------\n')

for characters in strings:
    print(characters)
    
print('----------- Dictionary ---------------\n')

for i in dictionary:
    print(i)
print('--------- Value Items ----------------\n')

for key, value in dictionary.items():
    print(key, value)
print('--------- Key Values -----------------\n')

for value in dictionary.values():
    print(value)

```
