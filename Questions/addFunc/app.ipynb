# Add Numbers App
This notebook demonstrates importing and using the `addNumber` module.

```python
# Import the addNumber module
import addNumber
```

```python
# Add two numbers
result = addNumber.add(5, 3)
print(f"5 + 3 = {result}")
```

```python
# Add multiple numbers
result_multiple = addNumber.add_multiple(1, 2, 3, 4, 5)
print(f"1 + 2 + 3 + 4 + 5 = {result_multiple}")
```

```python
# Try with floats
result_float = addNumber.add(3.14, 2.86)
print(f"3.14 + 2.86 = {result_float}")
```
